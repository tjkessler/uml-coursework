```python
from rsa.miller_rabin import generate_key
from rsa.core import encrypt_hex, decrypt_hex

public_key, private_key = generate_key(512, 40)

msg = 'Hello, World!'.encode('utf8').hex()
ctxt = encrypt_hex(msg, public_key)
dmsg = decrypt_hex(ctxt, private_key)

print(msg == dmsg)
```

```python
from string import ascii_letters
from random import choices
from time import time
from matplotlib import pyplot as plt

n_letters = []
times = []
for i in range(1, 128):
    msg = ''.join(choices(ascii_letters, k=i)).encode('utf8').hex()
    _t_start = time()
    _ = encrypt_hex(msg, public_key)
    times.append(time() - _t_start)
    n_letters.append(i)

plt.clf()
plt.rcParams['font.family'] = 'Times New Roman'
plt.xlabel('Length of message, in ASCII characters')
plt.ylabel('Time to encrypt message w/ 512-bit public key (s)')
plt.plot(n_letters, times, color=(0, 114/235, 178/235))
plt.show()
```
