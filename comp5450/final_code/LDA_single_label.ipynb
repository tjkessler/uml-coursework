```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import OrdinalEncoder

trg_file = 'subset_train_expanded.csv'
tst_file = 'subset_test_expanded.csv'
#val_file = 'subset_valid_expanded.csv'


#Loading training subset of dataset
trg_ds = pd.read_csv(trg_file)
tst_ds = pd.read_csv(tst_file)
#val_ds = pd.read_csv(val_file)
```

```python
#Lists identifying the features and labels
features = ['Type','Air temperature [K]', 'Process temperature [K]', 'Rotational speed [rpm]','Torque [Nm]','Tool wear [min]']
tgts = ['Machine failure','TWF','HDF','PWF','OSF','RNF']
```

```python
#Feature vector and labels for training data. !!!Only checking for total machine failures for now
X_trg = trg_ds[features]
y_trg_MF = trg_ds['Machine failure']
y_trg_TWF = trg_ds['TWF']
y_trg_HDF = trg_ds['HDF']
y_trg_PWF = trg_ds['PWF']
y_trg_OSF = trg_ds['OSF']
y_trg_RNF = trg_ds['RNF']

#Feature vector and labels for test data
X_tst = tst_ds[features]
y_tst_MF = tst_ds['Machine failure']
y_tst_TWF = tst_ds['TWF']
y_tst_HDF = tst_ds['HDF']
y_tst_PWF = tst_ds['PWF']
y_tst_OSF = tst_ds['OSF']
y_tst_RNF = tst_ds['RNF']


###############################
#Finding categorical variables (Here it's only "Type")
s = (X_trg.dtypes == 'object')
object_cols = list(s[s].index)


# Apply ordinal encoder to object columns
ordinal_encoder = OrdinalEncoder()

X_trg[object_cols] = ordinal_encoder.fit_transform(X_trg[object_cols])
X_tst[object_cols] = ordinal_encoder.transform(X_tst[object_cols])
```

```python
#Declaring classifiers for each failure type
#Was also implemented more concisely with MultiOutputClassifier object in scikitlearn, 
#which does essentially the same thing at the back end, but as one of the other classifiers was done with a different
#library that was not compatible with MultiOutputClassifer, the following implementation was carried out for
#a more valid comparison across models

clf_LDA_MF = LinearDiscriminantAnalysis()
clf_LDA_TWF = LinearDiscriminantAnalysis()
clf_LDA_HDF = LinearDiscriminantAnalysis()
clf_LDA_PWF = LinearDiscriminantAnalysis()
clf_LDA_OSF = LinearDiscriminantAnalysis()
clf_LDA_RNF = LinearDiscriminantAnalysis()
```

```python
#Training/Fitting all the classifiers
clf_LDA_MF.fit(X_trg, y_trg_MF)
clf_LDA_TWF.fit(X_trg, y_trg_TWF)
clf_LDA_HDF.fit(X_trg, y_trg_HDF)
clf_LDA_PWF.fit(X_trg, y_trg_PWF)
clf_LDA_OSF.fit(X_trg, y_trg_OSF)
clf_LDA_RNF.fit(X_trg, y_trg_RNF)
```

```python
#Making predictions for all the classifiers
pred_y_tst_MF = clf_LDA_MF.predict(X_tst)
pred_y_tst_TWF = clf_LDA_TWF.predict(X_tst)
pred_y_tst_HDF = clf_LDA_HDF.predict(X_tst)
pred_y_tst_PWF = clf_LDA_PWF.predict(X_tst)
pred_y_tst_OSF = clf_LDA_OSF.predict(X_tst)
pred_y_tst_RNF = clf_LDA_RNF.predict(X_tst)
```

```python
#Confusion Matrices over the test set
cm_tst = confusion_matrix(y_tst_MF, pred_y_tst_MF)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_tst).plot()
plt.title('Machine Failure (MF)')
plt.show()

cm_tst = confusion_matrix(y_tst_TWF, pred_y_tst_TWF)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_tst).plot()
plt.title('Tool Wear Failure (TWF)')
plt.show()

cm_tst = confusion_matrix(y_tst_HDF, pred_y_tst_HDF)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_tst).plot()
plt.title('Heat Dissipation Failure(HDF)')
plt.show()

cm_tst = confusion_matrix(y_tst_PWF, pred_y_tst_PWF)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_tst).plot()
plt.title('Power Failure (PWF)')
plt.show()

cm_tst = confusion_matrix(y_tst_OSF, pred_y_tst_OSF)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_tst).plot()
plt.title('Overstrain Failure (OSF)')
plt.show()

cm_tst = confusion_matrix(y_tst_RNF, pred_y_tst_RNF)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_tst).plot()
plt.title('Random Failure (RNF)')
plt.show()
```

```python
print('Classification Report for MF Classifier:\n')
print(classification_report(y_tst_MF, pred_y_tst_MF))
```
